```python
import polars
import hvplot.polars

import holoviews as hv

from polars import col

HEIGHT = 600
WIDTH = 1200
```

```python
amd = polars.read_csv("../amd.csv", try_parse_dates=True)
amd.head()
```

```python
ohlc = amd.hvplot.ohlc(x="Date", y=["Open", "High", "Low", "Close"], grid=True, width=WIDTH, height=HEIGHT)
vol = amd.select(polars.col("Date", "Volume")).hvplot.bar(x="Date", y="Volume", grid=True, width=WIDTH, height=int(HEIGHT / 3))

hv.Layout(ohlc + vol).cols(1)
```
