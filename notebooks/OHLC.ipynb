```python
import polars
import hvplot.polars

import holoviews as hv

from polars import col

HEIGHT = 600
WIDTH = 1200
```

```python
ohlc = amd.tail(50).hvplot.ohlc(x="Date", y=["Open", "High", "Low", "Close"], grid=True, width=WIDTH, height=HEIGHT)
ohlc
```
