```python
import polars
import hvplot.polars

import holoviews as hv
import utils

from polars import col
```

```python
amd = utils.load_asset_data("AMD").tail(50)
amd.head()
```

```python
import indicatory.ranges_returns as rr
```

```python
amd_daily_range = rr.daily_range(amd)
amd_daily_returns = rr.daily_returns(amd)
amd_atr = rr.average_true_range(amd, 10)
```

```python
ohlc = utils.ohlc_plot(amd)
range_plot = utils.bar_plot(amd_daily_range, value_field=rr.daily_range_percent_column())
returns_plot = utils.bar_plot(amd_daily_returns, value_field=rr.daily_returns_column(), color="red")

hv.Layout(ohlc + range_plot + returns_plot).cols(1)
```

```python
ohlc = utils.ohlc_plot(amd.tail(40))
atr_plot = utils.bar_plot(amd_atr.tail(40), value_field=rr.atr_percent_column(10))

hv.Layout(ohlc + atr_plot).cols(1)
```
