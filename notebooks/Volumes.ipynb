```python
import polars
import hvplot.polars

import holoviews as hv

from polars import col

import utils
```

```python
amd = utils.load_asset_data("AMD").tail(50)
amd.head()
```

```python
import indicatory.volumes as vol
```

```python
amd_obv = vol.on_balance_volume(amd)
```

```python
ohlc = utils.ohlc_plot(amd)
volume_plot = utils.bar_plot(amd.select(polars.col("Date", "Volume")), value_field="Volume")

obv_bars = utils.bar_plot(amd_obv, value_field="OBV", color="orange")
# obv_line = utils.lines_plot(amd_obv, value_fields=["OBV"])

hv.Layout(ohlc + volume_plot + obv_bars).cols(1)
```

```python
amd_pv = vol.price_by_volume(amd)
```

```python
amd_pv_plot = utils.bar_plot(amd_pv, value_field=vol.pv_column(), color="orange")

hv.Layout(ohlc + volume_plot + amd_pv_plot).cols(1)
```
